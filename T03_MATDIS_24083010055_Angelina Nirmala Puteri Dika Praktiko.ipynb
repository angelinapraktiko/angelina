##n^2-n=n(n-1)selalu dapat dibagi habis oleh (n-1)

```python
def cek_pembagian(n):
    expr = n**2 - n
    divisor = n - 1
    if divisor != 0:
        hasil = expr // divisor
        divisible = expr % divisor == 0
    else:
        hasil = "tak terdefinisi"   # karena n-1 = 0, tidak bisa dibagi
        divisible = False
    return expr, divisor, hasil, divisible

print(f"{'n':>3} {'n^2-n':>10} {'n-1':>10} {'(n^2-n)/(n-1)':>15} {'Habis dibagi?':>15}")
for n in range(-5, 21):
    expr, divisor, hasil, divisible = cek_pembagian(n)
    print(f"{n:3} {expr:10} {divisor:10} {str(hasil):>15} {str(divisible):>15}")
```

Hasil output membuktikan jika nilai n selai 1, hasil pembagian selalu bernilai sama dengan n. Namun pada n=1 menghasilkan nilai tak terdefinisi karena penyebut berasal dari (n-1)=0, sehingga pembagian menjadi tak terdefinisi. Kesimpulannya adalah n^2 - n selalu habis dibagi oleh n-1 untuk semua bilangan bulat kecuali n=1

```python
import numpy as np
import matplotlib.pyplot as plt

# rentang n
n_values = np.arange(-5, 21)

# hitung hasil (n^2 - n) / (n - 1), kecuali untuk n=1
hasil = []
for n in n_values:
    if n != 1:
        hasil.append((n**2 - n) / (n - 1))
    else:
        hasil.append(np.nan)  # NaN karena tidak terdefinisi

# plot grafik
plt.figure(figsize=(8,5))
plt.plot(n_values, hasil, marker='o', label=r"$(n^2-n)/(n-1)$")
plt.axvline(1, color='red', linestyle='--', label="n=1 (undef)")

plt.title("Grafik $(n^2-n)/(n-1)$ terhadap n")
plt.xlabel("n")
plt.ylabel("Hasil pembagian")
plt.grid(True)
plt.legend()
plt.show()
```

Grafik di atas menunjukkan hubungan antara nilai n dan hasil pembagian n^2-n/n-1. Untuk semua n≠1, grafik menghasilkan titik-titik yang terletak pada garis lurus y=n. Pada n=1 terdapat garis putus-putus merah yang menandakan bahwa nilai tersebut tak terdefinisi, karena pembagiannya adalah nol, sehingga hasilnya tidak dapat ditentukan.

##1 + 1/2 + 1/3 + ... + 1/n < 2

```python
def harmonic_series(n):
    total = 0
    for i in range(1, n + 1):
        total += 1 / i
    return total

print(f"{'n':>3} | {'H_n (sum)':>15} | {'H_n < 2?'}")
print("-" * 35)

for n in range(1, 21):  # bisa diganti range sesuai kebutuhan
    H_n = harmonic_series(n)
    print(f"{n:3} | {H_n:15.6f} | {H_n < 2}")
```

Untuk n=1, n=2, n=3, nilai Hn masih lebih kecil dari 2, masing-masing sebesar 1.000000, 1.500000, 1.833333. Sehingga memenuhi ketidaksamaan Hn<2. Namun, mulai dari n=4, nilai deret harmonik mencapai 2.083333 dan terus bertambah. Dengan demikian, syarat Hn<2 hanya berlaku pada tiga suku pertama, sedangkan n≥4, niai Hn selalu lebih besar dari 2 dan terus meningkat seiring pertambahan n. Hal ini sjealan dnegan isfat deret harmonik yang divergen, yaitu niainya bertambah tak terbats walaupun pertambahannya semakin kecil

```python
import matplotlib.pyplot as plt

# Fungsi deret harmonik
def harmonic_series(n):
    return sum(1/i for i in range(1, n+1))

# Data
n_values = list(range(1, 21))
H_values = [harmonic_series(n) for n in n_values]

# Plot
plt.figure(figsize=(8,5))
plt.plot(n_values, H_values, marker='o', label='Harmonic Series H_n')
plt.axhline(y=2, color='r', linestyle='--', label='y = 2')

plt.title("Grafik Deret Harmonik H_n")
plt.xlabel("n")
plt.ylabel("H_n")
plt.legend()
plt.grid(True)
plt.show()
```

Kurva Hn terus eningkat seiring bertambahanya n, meskipun kenaikan yang terjadi semakin kecil pada nilai n yang lebih besar. Garis horizontal merah menunjukkanbatas y=2. Grafik n=3, nilai Hn masih berada di bawah garis batas tersebut, tetapi mulai pada n=4, kurva Hn melampaui garis y=2 dan terus naik
