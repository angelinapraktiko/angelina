##Frekuensi


```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PyPDF2 import PdfReader



file_path = "/content/82707-235650-1-SM.pdf"
reader = PdfReader(file_path)
text = ""
for page in reader.pages:
    text += page.extract_text()



pattern = (
    r"Abstrak:(.*?)"
    r"1\.\s*PENDAHULUAN(.*?)"
    r"5\.\s*KESIMPULAN DAN SARAN(.*?)"
    r"(DAFTAR PUSTAKA|References)"
)
match = re.search(pattern, text, re.S | re.I)
if match:
    selected_text = " ".join(match.groups()[:-1])
    print("Bagian teks ditemukan.")
else:
    print("Bagian teks tidak ditemukan.")
    selected_text = ""



clean_text = re.sub(r"[^A-Za-z-\s]", " ", selected_text.lower())
clean_text = re.sub(r"\s+", " ", clean_text).strip()



words = re.findall(r"[A-Za-z-]+", clean_text)
vowels = set("aeiou")

def count_vowels_consonants(word):
    v = sum(1 for c in word if c in vowels)
    k = sum(1 for c in word if c.isalpha() and c not in vowels)
    return v, k



counts_all = [count_vowels_consonants(w) for w in words]
df_all = pd.DataFrame(counts_all, columns=["vokal", "konsonan"])
freq_table = df_all.value_counts().reset_index(name="frekuensi")



pivot_freq = freq_table.pivot_table(
    index="vokal", columns="konsonan", values="frekuensi", fill_value=0
)


if not pivot_freq.empty:
    plt.figure(figsize=(10,6))
    sns.heatmap(pivot_freq, annot=True, fmt=".0f", cmap="YlGnBu")
    plt.title("Heatmap Frekuensi Kombinasi VokalKonsonan ")
    plt.xlabel("Jumlah Konsonan")
    plt.ylabel("Jumlah Vokal")
    plt.show()
else:
    print("Tidak ada data untuk membuat heatmap.")
```

Berdasarkan hasil hetamap, terlihat bahwa analisis ini meanmpilkan distribusi frekuensi berbagai kombinasi jumlah huruf vokal dan konsonan pada bagain abstrak, Pendahuluan, serta Kesimpulan dan Saran. Kombinasi dengan 2-3 vokal dan 2-4 konsonan muncul paling sering dengan frekuensi mencapai ratusan kali, sedangkan kombinasi di luar rentang tersebut jauh lebih jarang atau bahkan tidak ditemukan. Secara umum, hasil ini menunjukkan bahwa kata dalam teks cenderung memiliki struktur huruf yang seimbang dengan panjang sedang, sehingga membentuk pola distribusi yang terpusat pada kombinasi vokal-konsonan tersebut

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PyPDF2 import PdfReader

# === 1. Baca teks dari PDF ===
file_path = "/content/82707-235650-1-SM.pdf"
reader = PdfReader(file_path)
text = ""
for page in reader.pages:
    text += page.extract_text()

# === 2. Ekstraksi bagian Abstrak, Pendahuluan, dan Kesimpulan ===
pattern = (
    r"Abstrak:(.*?)"
    r"1\.\s*PENDAHULUAN(.*?)"
    r"5\.\s*KESIMPULAN DAN SARAN(.*?)"
    r"(DAFTAR PUSTAKA|References)"
)
match = re.search(pattern, text, re.S | re.I)

if match:
    selected_text = " ".join(match.groups()[:-1])
    print("Bagian teks ditemukan.")
else:
    print("Bagian teks tidak ditemukan.")
    selected_text = ""

# === 3. Bersihkan teks ===
clean_text = re.sub(r"[^A-Za-z-\s]", " ", selected_text.lower())
clean_text = re.sub(r"\s+", " ", clean_text).strip()

# === 4. Pisahkan kata ===
words = re.findall(r"[A-Za-z-]+", clean_text)
vowels = set("aeiou")

# === 5. Hitung jumlah vokal dan konsonan per kata ===
def count_vowels_consonants(word):
    v = sum(1 for c in word if c in vowels)
    k = sum(1 for c in word if c.isalpha() and c not in vowels)
    return v, k

counts_all = [count_vowels_consonants(w) for w in words]
df_all = pd.DataFrame(counts_all, columns=["vokal", "konsonan"])
freq_table = df_all.value_counts().reset_index(name="frekuensi")

# === 6. Bubble Plot Frekuensi Kombinasi VokalKonsonan ===
if not freq_table.empty:
    plt.figure(figsize=(8,6))
    plt.scatter(
        freq_table["konsonan"],
        freq_table["vokal"],
        s=freq_table["frekuensi"] * 50,   # ukuran gelembung proporsional dengan frekuensi
        c=freq_table["frekuensi"],
        cmap="viridis",
        alpha=0.7,
        edgecolor="black"
    )
    plt.colorbar(label="Frekuensi")
    plt.title("Bubble Plot Frekuensi Kombinasi VokalKonsonan")
    plt.xlabel("Jumlah Konsonan")
    plt.ylabel("Jumlah Vokal")
    plt.grid(True, linestyle="--", alpha=0.4)
    plt.show()
else:
    print("Tidak ada data untuk membuat bubble plot.")
```

Visualisasi bubble plot memperlihatkan bahwa sebagian besar kata dalam teks memiliki 1-3 vokal dan 2-4 konsonan. Ukuran dan warna gelembung yang lebih besar di area tersebut menunjukkan bahwa kombinasi ini paling sering muncul, sedangkan kata dengan jumlah huruf yang lebih banyak muncul lebih jarang

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PyPDF2 import PdfReader


file_path = "/content/82707-235650-1-SM.pdf"
reader = PdfReader(file_path)
text = ""
for page in reader.pages:
    text += page.extract_text()



pattern = (
    r"Abstrak:(.*?)"
    r"1\.\s*PENDAHULUAN(.*?)"
    r"5\.\s*KESIMPULAN DAN SARAN(.*?)"
    r"(DAFTAR PUSTAKA|References)"
)
match = re.search(pattern, text, re.S | re.I)
if match:
    selected_text = " ".join(match.groups()[:-1])
    print("Bagian teks ditemukan.")
else:
    print("Bagian teks tidak ditemukan.")
    selected_text = ""



clean_text = re.sub(r"[^A-Za-z-\s]", " ", selected_text.lower())
clean_text = re.sub(r"\s+", " ", clean_text).strip()



words = re.findall(r"[A-Za-z-]+", clean_text)
unique_words = set(words)
vowels = set("aeiou")

def count_vowels_consonants(word):
    v = sum(1 for c in word if c in vowels)
    k = sum(1 for c in word if c.isalpha() and c not in vowels)
    return v, k



counts_unique = [count_vowels_consonants(w) for w in unique_words]
df_unique = pd.DataFrame(counts_unique, columns=["vokal", "konsonan"])
count_table = df_unique.value_counts().reset_index(name="cacah_kata")



pivot_count = count_table.pivot_table(
    index="vokal", columns="konsonan", values="cacah_kata", fill_value=0
)



if not pivot_count.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(pivot_count, annot=True, fmt=".0f", cmap="PuBuGn")
    plt.title("Heatmap Cacah Kata Unik per Kombinasi VokalKonsonan")
    plt.xlabel("Jumlah Konsonan")
    plt.ylabel("Jumlah Vokal")
    plt.show()
else:
    print("Tidak ada data untuk membuat heatmap.")
```

Berdasarkan hasil visualisasi heatmap, terlihat bahwa cacah kata unik tersebar berbeda untuk setiap kombinasi jumlah vokal dan konsonan. Kombinasi dengan 2-3 vokal dan 2-4 konsonan memiliki jumlah kata unik paling banyak, sedangkan kombinasi dengan jumlah huruf yang terlalu sedikit atau terlalu banyak muncul lebih jarang. Hal ini menunjukkan bahwa kata unik dalam teks umumnya memiliki struktur huruf yang seimbang dan panjang yang sedang.

```python
# --- Bubble Plot Cacah Kata Unik ---
plt.figure(figsize=(8, 6))
plt.scatter(
    count_table["konsonan"],
    count_table["vokal"],
    s=count_table["cacah_kata"] * 50,  # ubah angka kalau mau gelembung lebih kecil/besar
    c=count_table["cacah_kata"],
    cmap="YlOrBr",
    alpha=0.7,
    edgecolor="black"
)


plt.colorbar(label="Cacah Kata Unik")
plt.title("Bubble Plot Cacah Kata Unik per Kombinasi VokalKonsonan")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

#Jumlah kata dalam kalimat

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from PyPDF2 import PdfReader


file_path = "/content/82707-235650-1-SM.pdf"
reader = PdfReader(file_path)
text = ""
for page in reader.pages:
    text += page.extract_text()


pattern = (
    r"Abstrak:(.*?)"
    r"1\.\s*PENDAHULUAN(.*?)"
    r"5\.\s*KESIMPULAN DAN SARAN(.*?)"
    r"(DAFTAR PUSTAKA|References)"
)
match = re.search(pattern, text, re.S | re.I)

if match:
    selected_text = " ".join(match.groups()[:-1])
    print("Bagian teks ditemukan.")
else:
    print("Bagian teks tidak ditemukan, pastikan format PDF sesuai.")
    selected_text = ""



clean_text = re.sub(r"[^A-Za-z-\s.!?]", " ", selected_text)
clean_text = re.sub(r"\s+", " ", clean_text).strip()



sentences = re.split(r'(?<=[.!?])\s+', clean_text)
sentences = [s.strip() for s in sentences if s.strip()]



word_counts = [len(re.findall(r"[A-Za-z-]+", s)) for s in sentences]



df_sentences = pd.DataFrame({
    "Kalimat": sentences,
    "Jumlah_Kata": word_counts
})



plt.figure(figsize=(10, 6))
plt.hist(word_counts, bins=8, color="#5DADE2", edgecolor='black', alpha=0.8)
plt.title("Distribusi Jumlah Kata per Kalimat\n(Abstrak, Pendahuluan, dan Kesimpulan & Saran)",
          fontsize=13, fontweight='bold')
plt.xlabel("Jumlah Kata dalam Satu Kalimat", fontsize=11)
plt.ylabel("Frekuensi (Jumlah Kalimat)", fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()



print("=== Statistik Jumlah Kata per Kalimat ===")
print(df_sentences["Jumlah_Kata"].describe().round(2))


```

Berdasarkan hasil visualisasi histogram, distribusi jumlah kata per kalimta menunjukkan bahwa sebagian besar kalimat dalam teks memiliki panjang sekitar 16-28 kata. Nilai rata-rata jumlah per kalimat adalah 21, dengan median juga 21 kata, yang menandakan panjang kalimat cukup konsisten di sekitar niali tersebut. Kalimat terpanjang memiliki 66 kata. Secara umum, hasil ini menggambarkan bahwa struktur kalimat dalam teks cenderung sedang tidak terlalu pendek dan tidak terlalu panjang.

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from PyPDF2 import PdfReader
import numpy as np


file_path = "/content/82707-235650-1-SM.pdf"
reader = PdfReader(file_path)
text = ""
for page in reader.pages:
    content = page.extract_text()
    if content:
        text += content + "\n"



pattern = (
    r"Abstrak:(.*?)"
    r"1\.\s*PENDAHULUAN(.*?)"
    r"5\.\s*KESIMPULAN DAN SARAN(.*?)"
    r"(DAFTAR PUSTAKA|References)"
)
match = re.search(pattern, text, re.S | re.I)
if match:
    selected_text = " ".join(match.groups()[:-1])
    print("Bagian teks ditemukan.")
else:
    print("Bagian teks tidak ditemukan.")
    selected_text = ""



clean_text = re.sub(r"[^A-Za-z-\s.!?\n]", " ", selected_text)
clean_text = re.sub(r"\s+", " ", clean_text).strip()



paragraphs = re.split(r'(?<=[.!?])\s+(?=[A-Z])', clean_text)
paragraphs = [p.strip() for p in paragraphs if len(p.strip()) > 0]



def count_sentences(paragraph):
    sentences = re.split(r'(?<=[.!?])\s+', paragraph)
    sentences = [s.strip() for s in sentences if len(s.strip()) > 0]
    return len(sentences)

sentence_counts = [count_sentences(p) for p in paragraphs]



df_paragraphs = pd.DataFrame({
    "Nomor_Paragraf": range(1, len(paragraphs) + 1),
    "Jumlah_Kalimat": sentence_counts
})



plt.figure(figsize=(10, 6))

num_bins = min(10, len(set(sentence_counts))) if len(sentence_counts) > 1 else 1
plt.hist(df_paragraphs["Jumlah_Kalimat"],
         bins=np.linspace(min(sentence_counts)-0.5, max(sentence_counts)+0.5, num_bins+1),
         color="#48C9B0", edgecolor="black", alpha=0.85, rwidth=0.9)

plt.title("Distribusi Jumlah Kalimat per Paragraf\n(Abstrak, Pendahuluan, dan Kesimpulan & Saran)",
          fontsize=13, fontweight="bold")
plt.xlabel("Jumlah Kalimat dalam Paragraf", fontsize=11)
plt.ylabel("Frekuensi (Jumlah Paragraf)", fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()


counts, bins = np.histogram(df_paragraphs["Jumlah_Kalimat"], bins=num_bins)
for i, c in enumerate(counts):
    plt.text((bins[i] + bins[i+1]) / 2, c, str(int(c)),
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.show()



print("=== Statistik Jumlah Kalimat per Paragraf ===")
print(df_paragraphs["Jumlah_Kalimat"].describe().round(2))
```

 Berdasarkan hasil visualisasi histogram, sebagian besar paragraf dalam teks hanya terdiri dari suatu kalimat, dengan jumlah mencapai 79 paragraf. Rata-rata jumlah kalimat perparagraf adalah 1.355 dan nilai mediannya 1 yang menunjukkan bahwa struktur paragraf cenderung sangat singkat. Hanya seidkit paragraf yang memiliki lebih dari dua kalimat, dengan jumlah maksimal limat kalimat per paragraf. Secara umum, hasil ini menggabarkan bahwa teks disusun dengan paragraf paragraf pendek dan langsung pada inti pembahasan.

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from PyPDF2 import PdfReader


file_path = "/content/82707-235650-1-SM.pdf"
reader = PdfReader(file_path)
text = "".join([p.extract_text() for p in reader.pages if p.extract_text()])


pattern = (
    r"Abstrak:(.*?)"
    r"1\.\s*PENDAHULUAN(.*?)"
    r"5\.\s*KESIMPULAN DAN SARAN(.*?)"
    r"(DAFTAR PUSTAKA|References)"
)
match = re.search(pattern, text, re.S | re.I)
selected_text = " ".join(match.groups()[:-1]) if match else ""
print("Bagian teks ditemukan." if match else "Bagian teks tidak ditemukan.")


clean_text = re.sub(r"[^A-Za-z-\s.!?]", " ", selected_text)
clean_text = re.sub(r"\s+", " ", clean_text).strip()


paragraphs = re.split(r'(?<=[.!?])\s+(?=[A-Z])', clean_text)
paragraphs = [p.strip() for p in paragraphs if len(p.strip()) > 0]


def count_sentences(paragraph):
    sentences = re.split(r'(?<=[.!?])\s+', paragraph)
    sentences = [s.strip() for s in sentences if len(s.strip()) > 0]
    return len(sentences)

sentence_counts = [count_sentences(p) for p in paragraphs]


df = pd.DataFrame({
    "Nomor_Paragraf": range(1, len(paragraphs) + 1),
    "Jumlah_Kalimat": sentence_counts
})


plt.figure(figsize=(8, 4))
plt.boxplot(df["Jumlah_Kalimat"], vert=False, patch_artist=True,
            boxprops=dict(facecolor="#48C9B0", color="black", alpha=0.6),
            medianprops=dict(color="red", linewidth=2),
            whiskerprops=dict(color="black"),
            capprops=dict(color="black"),
            flierprops=dict(marker='o', markerfacecolor='orange', markersize=6, alpha=0.6))
plt.title("Sebaran Jumlah Kalimat per Paragraf", fontsize=13, fontweight="bold")
plt.xlabel("Jumlah Kalimat", fontsize=11)
plt.tight_layout()
plt.show()


print("=== Statistik Jumlah Kalimat per Paragraf ===")
print(df["Jumlah_Kalimat"].describe().round(2))
```

Hasil boxplot menunjukkan bahwa sebagain besar paragraf terdiri dari 1-2 kalimat, dengan median 1 kalimat. Hanya sedikit paragraf yang lbeih panjang, dengan max 5 kalimat.
