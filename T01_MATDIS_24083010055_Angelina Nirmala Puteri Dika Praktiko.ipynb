```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

df = pd.read_excel('/content/siswa.xlsx')
display(df)
```

```python
students = df["NAMA"].tolist()
X_matrix = df.drop(columns=["NPM", "NAMA"]).to_numpy()
```

```python
#Matriks Kovarian
cov_mat = np.cov(X_matrix, rowvar=False, ddof=1)
cov_df = pd.DataFrame(cov_mat, columns=df.columns[2:], index=df.columns[2:])


print("Matriks Kovarian")
print(cov_df, "\n")

# Heatmap Kovarian
plt.figure(figsize=(6,5))
sns.heatmap(cov_df, annot=True, cmap="Blues")
plt.title("Matriks Kovarian")
plt.show()
```

Berdasarkan hasil matriks kovarian, dapat dilihat jika setiap variabel memiliki tigkat variansi yang berbeda. Variabel suka membantu memiliki varians terbesar (0.0028), sedangkan variabel asyik, menarik memiliki varians terkecil (0.0016).

Terdapat hubungan yang cukup kuat antara variabel nyambung kalau ngobrol dengan variabel asyik, menarik (0.0015), hal ini menunjukkan bahwa siswa yang nyambung saat diajak berbincang cenderung dianggap asyik dan menarik.

Selain itu terdapat hubungan positif antara variabel adil, amanah dengan variabel asyik, menarik (0.0012). Sebaliknya, kovarians antara variabel adil, amanah dengan variabel suka membantu mendekati nol (0.00081).

```python
#Matriks Korelasi
corr_mat = np.corrcoef(X_matrix, rowvar=False)
corr_df = pd.DataFrame(corr_mat, columns=df.columns[2:], index=df.columns[2:])

print("Matriks Korelasi ===")
print(corr_df, "\n")

# Heatmap Korelasi
plt.figure(figsize=(6,5))
sns.heatmap(corr_df, annot=True, cmap="RdPu", vmin=-1, vmax=1)
plt.title("Matriks Korelasi")
plt.show()
```

Berdasarkan hasil matriks korelasi, terlihat jika sebagian besarhubungan antarvariabel memiliki nilai korelasi yang relatif rendah hingga sedang. Variabel nyambug kalau ngobrol memiliki korelasi sedang dengan variabel asyik, menarik (0.81), hal ini menunjukkan jika siswa yang mudah diajak berbincang atau mengobrol cenderung juga asyik dan menarik.

Selain itu, variabel adil, amanah juga memiliki korelasi positif dengan variabel asyik, menarik (0.58), hal ini menunjukkan bahwa siswa yang dipandang adil dan amanah cenderung dilihat lebih menarik.

Sementara itu, korelasi variabel suka membantu dengan variabel lainnya rendah (0.3 - 0.46), hal ini berarti variabel suka membantu tidak terllau berkaitan langsung dengan keadilan, amanah, dan kesan menarik. Nilai korelasi mendekati nol antara variabel adil, amanah dan suka membantu (0.3)

```python
#Cosine Similarity antar siswa
norms = np.linalg.norm(X_matrix, axis=1, keepdims=True)
cosine_sim = (X_matrix @ X_matrix.T) / (norms @ norms.T)
cosine_df = pd.DataFrame(cosine_sim, index=students, columns=students)

print("=== Cosine Similarity antar Siswa (5x5 pertama) ===")
print(cosine_df.iloc[:5, :5])

# Heatmap Cosine Similarity
plt.figure(figsize=(8,6))
sns.heatmap(cosine_df, cmap="YlOrRd")
plt.title("Cosine Similarity antar Siswa")
plt.show()
```

Cosine similarity pada heatmap  menggambarkan tingkat kemiripan pola jawaban antar siswa berdasarkan variabel yang dinilai. Nilai cosine similarity berkisar antara 0 hingga 1, dengan nilai mendekati 1 menunjukkan bahwa dua siswa memiliki pola jawaban yang sangat mirip, sedangkan nilai mendekati 0 menunjukkan bahwa pola jawaban mereka sangat berbeda.

Pada hasil visualisasi ini, hampir seluruh nilai berada sangat dekat dengan 1, yang ditunjukkan dengan dominasi warna merah gelap pada heatmap. Hal ini berarti sebagian besar siswa memiliki kecenderungan penilaian yang serupa terhadap aspek-aspek yang diukur, sehingga tidak terdapat perbedaan mencolok antar siswa.

Titik diagonal yang bernilai 1 menunjukkan perbandingan seorang siswa dengan dirinya sendiri. Sel-sel yang tampak sedikit lebih terang menandakan adanya perbedaan kecil dalam pola jawaban antar siswa tertentu.
