```python
import numpy as np
from scipy.integrate import cumulative_trapezoid
import matplotlib.pyplot as plt

# interval
x = np.linspace(0, 10, 500)

# fungsi sinusoidal
f = np.sin(x) + 0.5*np.sin(2*x) + (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)
```

##1. Hitung integral dan diferensial dari fungsi 'f' untuk interval x

```python
# turunan numerik
f_prime = np.gradient(f, x)

# integral kumulatif numerik (konstanta dibuat 0 di x[0])
F = cumulative_trapezoid(f, x, initial=0)

# contoh output
print("5 nilai pertama dari fungsi f(x)         :", f[:5])
print("5 nilai pertama turunan nurmeriknya f'(x):", f_prime[:5])
print("5 nilai pertama integral kumulatif F(x)  :", F[:5])
```

Hasil output menunjukkan lima nilai pertama dari masing-masing perhitungan.

Nilai fungsi f(x) dimulai dari 0 lalu meningkat bertahap hingga sekitar 0.396 yang menandakan fungsi awalnya naik

Turunan numerik f'(x) bernilai positif cukupbesar, sekitar 4.996 - 4.821, sehingga dapat disimpulkan fungsi sedang meningkat dnegan laju yang cukup tinggi, meskipun laju kenaikannya perlahan menurun

Integal Kumulatif F(x) dimulai dari nol lalu bertambah sedikit demi sedikit seiring bertambahnya nilai x

##2. Hitung secara terpisah bagian integral di atas dan dibawah y = **0**

```python
# pisahkan nilai positif & negatif
f_pos = np.where(f > 0, f, 0)
f_neg = np.where(f < 0, f, 0)

# luas (skalar) di area >0 dan <0
area_pos = np.trapezoid(f_pos, x)
area_neg = np.trapezoid(f_neg, x)

# integral kumulatif masing-masing
F_pos = cumulative_trapezoid(f_pos, x, initial=0)
F_neg = cumulative_trapezoid(f_neg, x, initial=0)

print("Luas di atas y=0 :", area_pos)
print("Luas di bawah y=0:", area_neg)
```

Hasil perhitungan menunjukkan jika luas area di atas sumbu y = 0 bernilai sekitar 4.604 yang berarti total area pada bagian fungsi f(x) yang positif lebih besar.

Sementara itu, luas area dibawah sumbu y = 0 bernilai sekitar -2.418. Jika hanya melihat besarannya saja, area negatif tersebut seluas 2.418, tetapi karena berada di bawah sumbu x, maka dianggap nilainya negatif.

##3. Plot semua fungsi: f, f', F (atas dan bawah dengan area warna berbeda)

```python
# turunan & integral kumulatif
f_prime = np.gradient(f, x)
F = cumulative_trapezoid(f, x, initial=0)

# untuk area & integral terpisah
f_pos = np.where(f > 0, f, 0)
f_neg = np.where(f < 0, f, 0)
F_pos = cumulative_trapezoid(f_pos, x, initial=0)
F_neg = cumulative_trapezoid(f_neg, x, initial=0)

plt.figure(figsize=(12, 8))

# f(x) + area
plt.subplot(3, 1, 1)
plt.plot(x, f, label='f(x)')
plt.axhline(0, linestyle='--', linewidth=1)
plt.fill_between(x, f, 0, where=f>0, alpha=0.4, label='Area > 0')
plt.fill_between(x, f, 0, where=f<0, alpha=0.4, label='Area < 0')
plt.title("f(x) dan area relatif terhadap y=0")
plt.legend()
plt.grid(alpha=0.3)

# f'(x)
plt.subplot(3, 1, 2)
plt.plot(x, f_prime, label="f'(x)")
plt.axhline(0, linestyle='--', linewidth=1)
plt.title("Turunan numerik f'(x)")
plt.legend()
plt.grid(alpha=0.3)

# F(x) + F_pos, F_neg
plt.subplot(3, 1, 3)
plt.plot(x, F, label='F(x) = ∫ f(x) dx')
plt.plot(x, F_pos, linestyle='--', label='∫ area positif')
plt.plot(x, F_neg, linestyle='--', label='∫ area negatif')
plt.title("Integral kumulatif (numerik)")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

 grafik pertama (f(x) dan area relatif terhadap y = 0). Area di atas sumbu x yang berwarna biru, menunjukkan bahwa nilai fungsi f(x) pada wilayah tersebut bernilai positif. Sebaliknya, area dibawah sumbu x yang berwarna oranye, menunjukkan bahwa nilai fungsi f(x) pada wilayah tersebut bernilai negatif


 Grafik kedua merupakan turunan numerik f'(x) dimana nilai positif menunukkan bahwa fungsi f(x) sedang meningkat, sedangkan nilai negatif menunjukkan fungsi sedang turun. Titik-titik dimana turunan bernilai nol biasanya merupakan puncak atau lembah dari f(x)


 Grafi ketiga merupakan visualisasi dari integral kumultaif F(x) = ∫ f(x) dx. Area positif lebih dominan, grafik F(x) naik, sedangkan jika area negatif lebih dominan, grafik F(x) turun. Selain itu diatmpilkan juga integral kumulatif area positif (Fpos) yang selalu ertambah, serta integral umulatif area negatif (Fneg) yang menurun karena hanya menghitung luas bagian di bawah sumbu x
