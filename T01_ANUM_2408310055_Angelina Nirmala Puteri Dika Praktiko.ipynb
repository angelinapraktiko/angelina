```python
# Nama : Angelina Nirmala Puteri Dika Praktiko
# NPM  : 24083010055

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# range harga jual
x = np.linspace(3000, 10000, 500)

# rescale parameters
a = 6
loc = 3.5 * 1000     # shift loc
scale = 1.5 * 1000   # widen scale

# probability density, right tail skew
y = skewnorm.pdf(x, a=a, loc=loc, scale=scale) * 1000

np.random.seed(55) # masukkan NPM
y = y + np.random.normal(0, 0.005, size=x.shape)  # add Gaussian noise
y = y*1000

# harga produksi
produksi_y= np.linspace(0, 1000, len(x))
produksi_z = np.linspace(3000, 2000, len(x)) + np.random.normal(0, 0.1, size=x.shape)*100
# linearly decreases from 3000 -> 2000 for quantity 0 -> 1000

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y, color="navy", label='penjualan', linewidth=2)
plt.plot(produksi_z, produksi_y, color="red", label='produksi', linewidth=2)
plt.title("Toko Roti Maknyus")
plt.xlabel("x (harga)")
plt.ylabel("y (jumlah barang dibuat/laku)")
plt.grid(True)
plt.legend()
plt.show()

# tugas:
# hitung volume produksi yang menghasilkan kentungan maksimum
#    (pendapatan dari penjualan dikurangi produksi)
# kirim ke muhammad_zulhaj.sada@upnjatim.ac.id
```

```python
# Nama : Angelina Nirmala Puteri Dika Praktiko
# NPM  : 24083010055

# Hitung Pendapatan (Revenue)
 # harga jual * jumlah barang laku
revenue = x * y

# Hitung Biaya Produksi (Cost)
 # harga produksi per unit * jumlah barang diproduksi
cost = produksi_z * produksi_y

# Hitung Keuntungan (Profit)
profit = revenue - cost

# Cari keuntungan maksimum
idx_max = np.argmax(profit)     # posisi keuntungan terbesar
q_optimal = produksi_y[idx_max] # volume produksi optimal (berapa banyak roti yang sebaiknya di produksi)
profit_max = profit[idx_max]    # keuntungan maksimum (nilai keuntungan terbesar yang bisa didapatkan)
harga_optimal = x[idx_max]      # harga jual optimal (harga jual per roti)

print(f"Volume produksi optimal : {q_optimal:.0f} unit")
print(f"Harga jual optimal      : Rp {harga_optimal:,.0f}")
print(f"Keuntungan maksimum     : Rp {profit_max:,.0f}")


plt.figure(figsize=(8,5))
plt.plot(produksi_y, profit, color="green", linewidth=2, label="Keuntungan")
plt.axvline(q_optimal, color="red", linestyle="-",
            label=f"Optimal q={q_optimal:.0f}")
plt.title("Keuntungan vs Volume Produksi")
plt.xlabel("Jumlah produksi (q)")
plt.ylabel("Keuntungan")
plt.legend()
plt.grid(True)
plt.show()
```

Perbedaan dari Grafik Pertama dan Kedua:
- Grafik Pertama -> fokus pada harga dan hubungannya dnegan produksi/penjualan
- Grafik kedua -> fokus pada jumlah produksi dan berapa besar keuntungan yang dihasilkan
